```python
#import matplot lib,numpy.
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
#getting input from user:
#joblist stores the job id's or number
#t_j denotes the duration time
#d_j denotes the deadline
#Do not enter 0
joblist =[];
t_j=[];
d_j=[];

n=int(input("enter the number of jobs"))
for i in range(0,n):
    print("Enter the duration time for",+i)
    t_j.append(int(input()));
    print("Enter the deadline time for",+i)
    d_j.append(int(input()));
    joblist.append(i);
length=len(d_j)
```

```python
#To show how the Greedy algorithm works, I am first trying to schedule the jobs in the order user inputs(users preference)
#and compare this schedule with schedule by greedy algorithm(sorting according to earliest deadline and scheduling)
#Each graph shows the schedule when a job is added along with the lateness it incured(shown beside the graph plot)
#At the end when all jobs are schedule in the order of user input ,the maximum is found and printed
#The interval(starttime and end time can be seen in x axis and y axis denotes the jobs)
fig, gnt=plt.subplots(n,1,figsize=(20,50))

a = 0;
y = 1;
arr = []
B_lateness = [];
Bmax_l = 0;
B_temp = [];
max_lb=0
l=0
ytick=[];
xtick=[];
xtick.append(0)
plt.title("Scheduling jobs based on the order in which user inputs")
for j in range(0, n):
    #gnt[j].set_ylim(0,(n*2))
    #gnt[j].set_xlim(0,50)
    gnt[j].grid(True)

    
    s_j = a;
    f_j = a + t_j[j]
    a = f_j;
    
    xtick.append(f_j)
    arr.append([s_j, t_j[j]])
    l=max(0, (f_j - d_j[j]))
    B_lateness.append(l)
    if (f_j > d_j[j]):
        max_lb = max(max_lb, (f_j - d_j[j]))
    B_temp.append((s_j, f_j, B_lateness[j]))
    gnt[j].broken_barh(arr, (y, 3), facecolor=('red', 'blue', 'black', 'pink', 'yellow', 'green','violet','cyan','magenta','olive','brown'))
    ytick.append(y);
    gnt[j].annotate(str(l),(f_j,y),xytext=(5, 0), textcoords="offset points",va='center')
    
    gnt[j].set_xticks(xtick)
    gnt[j].set_xlabel('time')
    gnt[j].set_ylabel('jobs')
    gnt[j].set_yticks(ytick)
    gnt[j].set_yticklabels(joblist)
    plt.figure()
    
    fig.show()
    
    y = y + 10;

print("The sequence of jobs before sorting by deadline")
plt.show()



for t in B_lateness:
    print(t)
print("the maximum lateness",+max_lb)
```

```python
#sorting the jobs according to the deadine(earliest deadline first)

for k in range(length):
    for l in range(0,n-k-1):
        if d_j[l] > d_j[l+1]:
            d_j[l], d_j[l+1] = d_j[l+1], d_j[l];
            t_j[l], t_j[l + 1] = t_j[l + 1], t_j[l];
            joblist[l], joblist[l + 1] = joblist[l + 1], joblist[l];
            print(d_j)
```

```python
for o in range(0,n):
    print("job",+joblist[o])
```

```python
#Greedy schedule
# schedule with earliest deadline first 
#Each graph shows the schedule when a job is added along with the lateness it incured(lateness shown beside the graph plot)
#At the end when all jobs are schedule in the order of user input ,the maximum is found and printed
#The interval(starttime and end time can be seen in x axis and y axis denotes the jobs)
#We can see that greedy schedule minimizes lateness
fig1, ax=plt.subplots(n,1,figsize=(20,50))

t = 0;
y = 1;
arr1 = []
A_lateness = [];
Amax_l = 0;
A_temp = [];
max_lA = 0
ytick_A=[];
xtick_A=[];
xtick_A.append(0)
plt.title("Greedy schedule for minimizing lateness(After sorting the jobs with deadine)")
print("Greedy schedule for minimizing lateness(After sorting the jobs with deadine)")
for j in range(0, n):
    #ax[j].set_ylim(0,100)
    #ax[j].set_xlim(0,50)
    ax[j].grid(True)

    s_j = t;
    f_j = t + t_j[j]
    t = f_j;
    xtick_A.append(f_j)
    arr1.append([s_j, t_j[j]])
    late=max(0, (f_j - d_j[j]))
    A_lateness.append(late)
    if (f_j > d_j[j]):
        max_lA = max(max_lA, (f_j - d_j[j]))
    A_temp.append((s_j, f_j, A_lateness[j]))
    ax[j].broken_barh(arr1, (y, 3), facecolor=('red', 'blue', 'black', 'pink', 'yellow', 'green','violet','cyan','magenta','olive','brown'))
    ytick_A.append(y);
    ax[j].annotate(str(late),(f_j,y),xytext=(5, 0), textcoords="offset points",va='center')
    ax[j].set_xticks(xtick_A)
    ax[j].set_ylabel('jobs')
    ax[j].set_yticks(ytick_A)
    ax[j].set_yticklabels(joblist)
    plt.figure()
    fig1.show()
    y = y + 10;


plt.show()
for t in A_lateness:
    print(t)
print("Maximum lateness",+max_lA)

```
